#### Chúng mình sẽ bắt đầu bằng một mô hình machine learning đơn giản.
Ở đây mình sẽ thử tạo ra 2 dãy số :

> x = -1, 0, 1, 2, 3, 4 <br> y = -3, -1, 1, 3, 5, 7

Như mọi người có thể thấy nhìn sơ qua chúng mình cũng đoán được quy luật của chúng là:
> y = 2x - 1

Mình sẽ tiến hành tạo một mô hình để học 2 dãy số trên và tiến hành dự đoán cho các số tiếp theo.

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

# Khởi tạo kiến trúc mô hình, ở dây mình chỉ sử dụng một lớp tuyến tính (Dense or Linear(Pytorch))
# units đại diện cho số lượng đầu ra của mình, ở đây mình chỉ muốn dự đoán kết quả y tức 1 số nên để là 1
# tương tự vậy với kích thước đầu vào là x, cũng là 1 số nên đẻ là 1. Nếu như x của bạn là 1 mảng (vector) thì input_shape chính là số lượng phần tử trong mảng đó
linear = Dense(units=1, input_shape=[1])

# Dùng Sequential để khởi tạo mô hình và đưa danh sách các lớp vào nha, ở đây mình chỉ dùng một lớp tuyến tính duy nhất.
model = Sequential([linear])

# Chọn hàm tối ưu (optimizer) là "Giảm dần độ dốc ngẫu nhiên" - SGD(Stochastic gradient descent)
# hàm mất mát (loss) là "Trung bình bình phương sai số " - MSE (Mean squared error)
model.compile(optimizer='sgd', loss='mean_squared_error')

# Khởi tạo dữ liệu cho việc huấn luyện
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# Huấn luyện mô hình với số epochs là 500
model.fit(xs, ys, epochs=500)
```

Ở đây mọi người có thể thấy 2 khái niệm khá mới với người mới học là hàm tối ưu và hàm mất mát ha. Chúng là 2 thứ hầu như sẽ xuất hiện xuyên suốt trong tất cả các mô hình, mình sẽ không đi sâu về toán nên là theo mình mọi người chỉ cần hiểu đơn giản là vầy:
- **Hàm mất mát** giúp mô hình **tính số lỗi sai** mà mình đã mắc phải, từ đó có thể nhận thức được kết quả từ quá trình học đã tốt hơn và giống với mục tiêu đề ra chưa. Qua đó làm tiền đề cho bước học tiếp theo.
- **Hàm tối ưu** giúp mô hình **tối ưu hóa kết quả, giảm thiểu số lỗi về mức thấp nhất** từ đó đạt được hiệu quả chính xác cao hơn. Nó giống như cách học của con người chúng ta vậy, con người có nhiều phương pháp, cách học khác nhau rồi luyện tập, sửa lỗi sai không ngừng cho đến khi đúng thì mô hình cũng thế.

```python
# Xuất kết quả dự đoán với x = 10
print("Kết quả dự đoán với x = 10 là:" ,model.predict(np.array([10.0])))
```

Bạn có thể thấy, tuy là không chính xác hoàn toàn nhưng với sai số chỉ nằm trong khoảng [0.01 - 0.02] thì đây là một con số khá ấn tượng.

Bây giờ tụi mình sẽ thử kiểm tra trọng số mà mô hình đã học được.

```python
print(f"Trọng số mô hình học được:\n{linear.get_weights()}")
```

Bạn có thể thấy nó gần giống với công thức ban đầu mà chúng mình xác định ra.

Công thức ban đầu:
> y = 2.x - 1

Công thức mô hình dự đoán (đây là kết quả trong lần chạy của mình, có thể ở mọi người sẽ ra số khác gần giống):
> y = 1,9969475.x - 0,9905362

### Vậy là kết thúc chương một rồi á, chỉ đơn giản là giới thiệu về cách cài đặt và sử dụng sơ các hàm trong tensorflow. Cảm ơn mọi người đã coi hết bài đầu tiên của mình.

Những kiến thức nắm bắt được:
- **Dense(units, input_shape)** để khởi tạo lớp tuyến tính ( linear).
- **Sequential([danh sách các lớp])** để khởi tạo kiến trúc mô hình.
- **model.compile(optimizer="hàm tối ưu", loss = "hàm mất mát")** để định nghĩa hàm tối ưu và hàm mất mát cho mô hình.
- **model.fit(dữ liệu train, dữ liệu test, số epochs)** để huấn luyện mô hình
- **model.predict(mảng cần dự đoán)** để lấy kết quả dự đoán của môi hình.
